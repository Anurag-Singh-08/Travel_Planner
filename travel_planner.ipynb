```python
#------------------
#import required libraries
#------------------

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# -----------------------------------------------------------
# Travel Destination Dataset (30 Global Locations)
# -----------------------------------------------------------


data = {
    "destination": [
        "Goa", "Manali", "Paris", "Dubai", "Bali",
        "New York", "Tokyo", "Switzerland", "Kerala", "Jaipur",
        "Maldives", "Singapore", "London", "Bangkok", "Kathmandu",
        "Sydney", "Cape Town", "Rio de Janeiro", "Istanbul", "Rome",
        "Hong Kong", "Barcelona", "Amsterdam", "Phuket", "Boracay",
        "Srinagar", "Leh Ladakh", "Andaman", "Vienna", "Seoul"
    ],
    "description": [
        "beaches nightlife water sports parties seafood",
        "mountains snow trekking adventure cold",
        "romance museums cafes architecture art culture",
        "luxury shopping deserts skyscrapers nightlife",
        "beaches temples surfing relaxation culture",
        "nightlife shopping skyscrapers museums food",
        "technology anime shopping culture food temples",
        "mountains lakes snow adventure luxury scenery",
        "backwaters nature relaxation greenery beaches",
        "palaces forts culture traditional food history",

        "beaches luxury relaxation snorkeling water villas",
        "shopping nightlife clean city futuristic attractions",
        "museums bridges cafes royal history architecture",
        "nightlife street food markets temples shopping",
        "mountains temples trekking spiritual nature",

        "beaches opera harbor surfing city life",
        "mountains beaches adventure wildlife vineyards",
        "beaches carnival nightlife mountains music",
        "mosques culture bazaars history architecture",
        "roman history architecture museums ancient culture",

        "skyscrapers shopping nightlife harbor food",
        "beaches architecture food nightlife culture",
        "canals museums cafes bicycles nightlife",
        "beaches parties nightlife water sports",
        "white sand beaches island nightlife relaxation",

        "lakes mountains houseboats snow gardens",
        "mountains desert adventure biking trekking",
        "islands beaches scuba diving blue water",
        "opera architecture classical music museums",
        "technology shopping nightlife kpop culture"
    ]
}

#The recommendation system uses these descriptions to match

df = pd.DataFrame(data)
df
```

```python
tfidf = TfidfVectorizer(stop_words="english")
tfidf_matrix = tfidf.fit_transform(df["description"])
tfidf_matrix.shape
```

```python
# Compute similarity matrix
similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

def recommend_destinations(user_interest, top_n=5):
    user_vec = tfidf.transform([user_interest])
    scores = cosine_similarity(user_vec, tfidf_matrix).flatten()
    top_indices = scores.argsort()[::-1][:top_n]
    return df.iloc[top_indices][["destination", "description"]]
```

```python
user_input = "I like beaches, nightlife and water sports"
recommend_destinations(user_input)
```

```python
recommend_destinations("I love mountains, snow and trekking")
```
